If we want to create a smoother plot without having to manually type hundreds of values into our array, we can generate an interval of evenly spaced numbers with the [`linspace`](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.linspace.html) function. VoilÃ , we just plotted a smooth sine function using only two lines of code! (Note: the plot is still using discrete values, only at a much higher resolution such that it appears to be continuous.)

For the following steps, we will use `numpy`'s random generator. To create reproducible results, we will now first set a seed such that the random generator always comes up with the same "random" numbers.

```python
np.random.seed(42)
```

Real data from the real world is never as perfect and clean as the sine wave above, there will always be noise in the data. We can simulate noisy data by adding random Gaussian noise to the clean sine wave like this:

```python
y = np.sin(x) + np.random.normal(loc=np.sin(x), scale=0.2)
plt.plot(x, y)
plt.show();
```

Play around with the `scale` parameter to see how it affects the signal. Head over to the [documentation](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.random.normal.html) for details about the function.

The most common task in Machine Learning is to find patterns in large amounts of noisy data. One of the simplest methods to find the underlying function of two-dimensional noisy data like this is to fit a polynomial function to the data. We can easily achieve this with NumPy's [`polyfit`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html) function. This function accepts a `deg` argument to control the polynomial degree of the fitted function. The degree is called a *hyperparameter* of the model, because it is set manually by the user. Our job is to find a degree that is high enough to model the data's complexity, but low enough to *not model the noise*.

The graph below compares polynomial models of different degrees to the original signal. You will see that if the degree is too low (2 and 3), the function cannot model the signal very well. Degrees 4 and up are indeed able to model the sine curve, however very high degrees (50) will start modeling the Gaussian noise of the signal.

```python
plt.figure(figsize=(16,9))
plt.plot(x, y, label="signal")
for deg in [2, 3, 4, 5, 15]:
    fit = np.poly1d(np.polyfit(x, y, deg))
    plt.plot(x, fit(x), label=deg, alpha=0.5, lw=2)
plt.legend()
plt.show();
```

Make sure you understand what's happening in the code cell above. Fiddle with the degree values to get an intuition of polynomial fitting, if this is new to you.

## Your first own learning algorithm: K-Means clustering

This demo shows a simple implementation of K-Means. Take a look at [Scikit-learn's User Guide](http://scikit-learn.org/stable/modules/clustering.html#k-means) for an introduction to the algorithm. Basically it comes down to the following steps:

- choose the number of clusters `k` and randomly initialize the cluster centers
- while clusters change do
    1. assign each data point to the nearest cluster center
    2. recompute the cluster centers as the mean of the cluster's assigned data points
    
Of course, the scikit-learn package comes with a [K-Means function](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html), which you can use in the end to verify your own results.

### Creating a dataset

For now, we'll set `k` to 3.
Let's start with generating a random set of 2D data points by using the [`make_blobs`](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) function:

```python
from sklearn.datasets import make_blobs
k = 3
X, Y = make_blobs(n_samples=800, n_features=2, centers=k, random_state=42, cluster_std=3)
cluster_colors = ["r", "g", "b"]
cluster_assignments = np.zeros(Y.shape, dtype=int)
plt.scatter(X[:,0], X[:,1], c=cluster_assignments, alpha=0.6)
plt.show();
```

We can already see in the plot that the data points roughly belong to 3 clusters. Change the following cell to initialize the three cluster centers at random coordinates within a sensible range. Hint: some [numpy random functions](https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html) allow you to set a (low, high) range from which values will be sampled.

```python
cluster_centers = np.random.randint(-15, 15, (k, 2))
cluster_centers
```

Next, fill in the code cell below such that it performs the iterative clustering as described in the steps above. Make sure the loop exits once the cluster changes do not change anymore. Use [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) for the cluster assignments, which you can implement yourself or use numpy's [`norm`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.norm.html) function.

Replace the `pass` statements with your code and add an `if` condition around changing the `cluster_changed` flag.

```python
cluster_changed = True
count = 0
while cluster_changed:
    print("iteration", count)
    # update cluster_assignments
    dist = np.vstack([np.linalg.norm(X - cluster_centers[i, :], axis=1) for i in range(k)])
    cluster_assignments = np.argmin(dist, axis=0)
    # recompute cluster centers
    prev_centers = np.copy(cluster_centers)
    for i in range(k):
        cluster_centers[i] = np.mean(X[cluster_assignments == i], axis=0)
    # check if clusters have changed
    cluster_changed = np.max(prev_centers - cluster_centers) > 0
    count += 1
```

After your loop has finished, execute the code below and check if the data points in the left are now clustered by color. The plot on the right is the ground truth for reference.

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.title("Your result")
plt.scatter(X[:,0], X[:,1], c=cluster_assignments, alpha=0.6)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker="*", s=400)
plt.subplot(1,2,2)
plt.title("Ground truth")
plt.scatter(X[:,0], X[:,1], c=Y, alpha=0.6)
plt.show();
```
